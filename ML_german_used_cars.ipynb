Correlation, Causation and ANOVA

Correlation is the measure of the interdepence between the variable. It return two values: pearson_coefficient and p-value

Causation is the relationship between the casue and effect between two variables.

ANOVA -> Analysis of Variance:  this is a statistical technique that is used to determine if there is significant difference between the average values (i.e  mean value) of two or more independent groups. It works by examining the spread of data within each group compared to the spread between the groups.

Anova return two values: f-test score and p-value

f-test sscore: Anova assumes all means of the groups are same, then calculate how much the actual mean deviates from the assumption. the calculated value is the f-test score. 

if our target column is strongly correlated with the variable we're analyzing, it is expected that the ANOVA test will return a large f_test score and a small p_value

Before going into prediction, we need to standardise.this is done by transforming data into a common format allowing us to make meaningful comparison.

Normaliasation - this the process of transofrming some columns the dataframe into similar range due to the high differences between the columns as it will affect our model. there is need to make the values be within range that will not make our prediction skew in a particular direction.
type of normalisation:
simple standardisation ->  scaling the variable so the average variable is 0
min-max standardisation -> scaling  the variable so the variance is 1
z-score standardisation -> scaling the variable so the variable vaues ranges from 0 to 1

In evaluating models we use regression plot and residual plot for for simple linear regression, while for multiple linear regression we use distribution plot

Regression plot show a combination of plots of scatter plot, showing the distribution of the data points and a line of best fit passing through the data points.

Residual polt is the type of plot that shows the variance of the data points. this graph shows the residuals on the y-axis and the target variable on the x-axis.
